```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del sistema
R = 1      # Ohm
L = 1      # H
C = 1      # F
Ve = 12    # Voltios (escaln)

# Parmetros derivados
zeta = R / (2 * np.sqrt(L / C))       # Factor de amortiguamiento
wn = 1 / np.sqrt(L * C)               # Frecuencia natural
wd = wn * np.sqrt(1 - zeta**2)        # Frecuencia amortiguada

# Vector de tiempo
t = np.linspace(0, 3, 1000)

# Soluciones analticas
i_t = Ve * (2 * np.sqrt(3) / 3) * np.exp(-t / 2) * np.sin((np.sqrt(3)/2) * t)
vc_t = Ve * (1 - (2 * np.sqrt(3) / 3) * np.exp(-t / 2) * np.sin((np.sqrt(3)/2) * t + np.pi/3))

# Grficos
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, i_t, 'b', label='i(t)')
plt.title('Corriente en el circuito RLC (respuesta analtica)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Corriente [A]')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, vc_t, 'r', label='V_c(t)')
plt.title('Tensin en el capacitor (respuesta analtica)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Tensin [V]')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
