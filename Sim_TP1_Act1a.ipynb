```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema
R = 1      # Ohm
L = 1      # H
C = 1      # F
Ve = 12    # Voltios (escalón)

# Parámetros derivados
zeta = R / (2 * np.sqrt(L / C))       # Factor de amortiguamiento
wn = 1 / np.sqrt(L * C)               # Frecuencia natural
wd = wn * np.sqrt(1 - zeta**2)        # Frecuencia amortiguada

# Vector de tiempo
t = np.linspace(0, 3, 1000)

# Soluciones analíticas
i_t = Ve * (2 * np.sqrt(3) / 3) * np.exp(-t / 2) * np.sin((np.sqrt(3)/2) * t)
vc_t = Ve * (1 - (2 * np.sqrt(3) / 3) * np.exp(-t / 2) * np.sin((np.sqrt(3)/2) * t + np.pi/3))

# Gráficos
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, i_t, 'b', label='i(t)')
plt.title('Corriente en el circuito RLC (respuesta analítica)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Corriente [A]')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, vc_t, 'r', label='V_c(t)')
plt.title('Tensión en el capacitor (respuesta analítica)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Tensión [V]')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
