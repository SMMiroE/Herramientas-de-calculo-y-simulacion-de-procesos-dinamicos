```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
R = 1
L = 1
C = 1
Ve = 12
t_final = 3
h = 0.001  # paso de integracin
N = int(t_final / h)

# Inicializacin
t = np.linspace(0, t_final, N+1)
i = np.zeros(N+1)
Vc = np.zeros(N+1)

# Condiciones iniciales
i[0] = 0
Vc[0] = 0

# Mtodo de Euler directo
for n in range(N):
    di_dt = (-R/L) * i[n] - (1/L) * Vc[n] + (1/L) * Ve
    dVc_dt = (1/C) * i[n]

    i[n+1] = i[n] + h * di_dt
    Vc[n+1] = Vc[n] + h * dVc_dt

# Grficos
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, i, label='i(t) numrica', color='blue')
plt.title('Corriente i(t) con Euler directo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Corriente [A]')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, Vc, label='Vc(t) numrica', color='red')
plt.title('Tensin en el capacitor Vc(t) con Euler directo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Tensin [V]')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
